Visualiser maintenant ces distances sur le plan, toujours avec matplotlib

```python
# Plotting the points and the distances
plt.figure(figsize=(10,8))

# Plotting the points
plt.scatter(points[:, 0], points[:, 1], color='blue')
for i, point in enumerate(points):
    plt.text(point[0]+0.2, point[1], f'P{i}', fontsize=10, ha='center')

# Plotting the distance lines and annotations
for i in range(num_points):
    for j in range(i+1, num_points):  # Avoid duplicate lines and self-connections
        plt.plot([points[i][0], points[j][0]], [points[i][1], points[j][1]], 'r--', alpha=0.5)
        midpoint = (points[i] + points[j]) / 2
        plt.text(midpoint[0], midpoint[1], f'{distances[i,j]:.2f}', fontsize=8, ha='center', color='red')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Distances entre les 10 Points Aléatoires')
plt.grid(True)
plt.show()
```
